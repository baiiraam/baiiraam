```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, StratifiedShuffleSplit
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
diabetes = pd.read_csv('/content/diabetes.csv')
```

```python
diabetes.head()
```

```python
X = diabetes.drop("Outcome", axis=1)
y = diabetes["Outcome"]

X, y
```

```python
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

X_scaled
```

```python
diabetes["Outcome"].value_counts()
```

```python
sss = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in sss.split(X, y):
  X_train, X_test = X.iloc[train_index], X.iloc[test_index]
  y_train, y_test = y.iloc[train_index], y.iloc[test_index]

X_scaled = scaler.transform(X_train)
X_scaled
```

```python
svm_clf = SVC()

svm_clf.fit(X_scaled, y_train)
y_pred = svm_clf.predict(X_scaled)

acc_sc_1 = accuracy_score(y_train, y_pred)
conf_mx = confusion_matrix(y_train, y_pred)
cl_rp_1 = classification_report(y_train, y_pred)

print(acc_sc_1)
print(conf_mx)
print(cl_rp_1)
```

```python
X_scaled_test = scaler.transform(X_test)
y_pred_test = svm_clf.predict(X_scaled_test)

acc_sc_2 = accuracy_score(y_test, y_pred_test)
conf_mx_2 = confusion_matrix(y_test, y_pred_test)
cl_rp_2 = classification_report(y_test, y_pred_test)

print(acc_sc_2)
print(conf_mx_2)
print(cl_rp_2)
```
